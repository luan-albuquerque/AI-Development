# Treinamento

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder
from sklearn.pipeline import Pipeline
```

```python
df_aluguel.idade_casa = df_aluguel.idade_casa.astype(int)
df_aluguel.tamanho_m2 = df_aluguel.tamanho_m2.astype(int)

df_aluguel.drop(df_aluguel[df_aluguel.idade_casa < 10].index, axis=0, inplace=True)
```

```python
df_aluguel.info()
```

```python
X = df_aluguel.drop(columns=['valor_aluguel'], axis=0)
y = df_aluguel['valor_aluguel']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3 , random_state=51)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model_reg = Pipeline(steps=[
    ('regressor', LinearRegression())
])
```

```python
model_reg.fit(X=X_train, y=y_train)
```

## Analise de Metricas

```python
y_pred = model_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error
```

```python
r2_score(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
root_mean_squared_error(y_test, y_pred)
```

# Residuos

```python
residuos = y_test - y_pred
```

```python
from scipy.stats import zscore

```

```python
residuos_std = zscore(residuos)
```

```python
sns.scatterplot(x=y_pred, y=residuos_std)
plt.axhline(y=0)
plt.axhline(y=-2)
plt.axhline(y=2)
```

```python
import pingouin as pg
plt.figure(figsize=(14,8))
pg.qqplot(residuos_std, dist='norm', confidence=0.95)
plt.xlabel('Quantis Teóricos')
plt.ylabel('Resíduos na escala padrão')
```

```python
# Realizar prediçoes individuais

predicao_individual = {
    'tamanho_m2': 106,
    'n_quartos' : 2,
    'idade_casa' : 39,
    'garagem' : 0,
    'localizacao_Periferia': True,
    'localizacao_Subúrbio' : False,
     
}

sample_df = pd.DataFrame(predicao_individual, index=[1])

model_reg.predict(sample_df)
```
